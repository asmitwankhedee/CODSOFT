Diving columns into X and Y

```python
new_col=df[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Cabin_Deck']]
new_col
```

```python
X=new_col.drop('Survived',axis=1)
y=new_col['Survived']
```

import libraries

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

 Model Train in 80% train and 20% for test

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

Use Random Forest model

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
y_predict
```

Confusion Metrix

```python
print(confusion_matrix(y_test,y_predict))
```

Classification Report to check Accuracy

```python
print(classification_report(y_test,y_predict))
```

Accuracy is 81%
